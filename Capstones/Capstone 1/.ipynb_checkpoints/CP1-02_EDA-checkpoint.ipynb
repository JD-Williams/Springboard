### Data Import

```python
# Import packages and modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# Assign file name
#file_path = 'data/clean_df.csv'
file_path = 'data/clean_comp_df.csv'

# Read CSV data into a Pandas dataframe
datetime_cols = ['incident_datetime','first_assignment_datetime',\
                 'first_activation_datetime','first_on_scene_datetime',\
                'first_to_hosp_datetime','first_hosp_arrival_datetime',\
                'incident_close_datetime']
df = pd.read_csv(file_path, parse_dates=datetime_cols,compression='gzip')
```

### Reshape and inspect data

```python
# Change dtypes
df['borough'] = df.borough.astype('category')
df['held_indicator'] = df.held_indicator.astype('category')
df['valid_dispatch_rspns_time_indc'] = df.valid_dispatch_rspns_time_indc.astype('category')
df['valid_incident_rspns_time_indc'] = df.valid_incident_rspns_time_indc.astype('category')
df['incident_dispatch_area'] = df.incident_dispatch_area.astype('category')
```

```python
# Set MultiIndex
df.set_index(['incident_year','incident_month','cad_incident_id'])
```

### Inicident Outcome

```python
df.patient_died.value_counts(normalize=True) # Rename 'patient_died' as 'fatality'
```

### Fatalitites in NYC (2013-2018)

```python
df[df.patient_died==True].groupby(['incident_year','incident_month']).patient_died.count() # Rename 'patient_died' as 'fatality'
```

```python
fatalities = df[df.patient_died==True].groupby(['incident_year','incident_month']).patient_died.count()
```

```python
fatalities.plot()
_ = plt.title('Fatalities Resulting from EMS Responses (2013-2018)')
_ = plt.ylabel('Count')
```

### Fatalities in NYC (by Final Call Type, 2013-2018)

```python
df[df.patient_died==True].final_call_type.value_counts().head(10)
```

```python
df[df.patient_died==True].final_call_type.value_counts().head().plot(kind='barh')
_ = plt.title('Top Five Call Types for Fatalities (2013-2018)')
_ = plt.xlabel('Count')
_ = plt.ylabel('Final Call Type')
```

```python
df[df.patient_died==True].final_call_type.value_counts(normalize=True).head(10)
```

### Fatalities in NYC (by Borough, 2013-2018)

```python
df[(df.patient_died==True) & (df.borough!='UNKNOWN')].borough.value_counts().head()
```

```python
df[(df.patient_died==True) & (df.borough!='UNKNOWN')].borough.value_counts().head().plot(kind='barh')
_ = plt.title('Fatalities in New York City (by Borough, 2013-2018)')
_ = plt.xlabel('Count')
_ = plt.ylabel('Borough')
```
