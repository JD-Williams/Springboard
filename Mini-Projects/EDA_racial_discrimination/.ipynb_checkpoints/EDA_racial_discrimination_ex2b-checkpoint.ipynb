#### Solution 1: Bootstrapping

```python
# Define a function that creates a bootstrap replicate
def bootstrap_replicate_1d(data, func):
    return func(np.random.choice(data, size=len(data)))

# Define a function to generate many bootstrap replicates
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data,func)
    return bs_replicates
```

#### Solution 2A: Compute statistics for each name group separately

```python
# Create samples for black and white callbacks, respectively
b = data[data.race=='b'].call
w = data[data.race=='w'].call

# Calculate mean and standard deviations for each sample
mu_b = np.mean(b)
std_b = np.std(b)
mu_w = np.mean(w)
std_w = np.std(w)

# Calculate the standard errors (se) for each group
n_b = len(b)
n_w = len(w)
se_b = std_b / np.sqrt(n_b)
se_w = std_w / np.sqrt(n_w)

# Calculate z-statistic per group
z_b = mu_b / se_b
z_w = mu_w / se_w

# Calculate p-value per group
pval_b = stats.norm.sf(z_b)
pval_w = stats.norm.sf(z_w)

# Calculate the margin of error (me) for a 95% C.I. per group
me_b = 1.96*se_b
me_w = 1.96*se_w

# Calculate the upper and lower bounds for each C.I.
upper_b = mu_b + me_b
lower_b = mu_b - me_b
upper_w = mu_w + me_w
lower_w = mu_w - me_w

# Print results per group
print('Callback rate for \'black-sounding names\': {:4f}'.format(mu_b))
print('p-value for black name sample: {}'.format(pval_b))
print('95% Confidence Interval for black name sample: {:4f} to {:4f}'.format(lower_b,upper_b))
print()
print('Callback rate for \'white-sounding names\': {:4f}'.format(mu_w))
print('p-value for white name sample: {}'.format(pval_w))
print('95% Confidence Interval for white name sample: {:4f} to {:4f}'.format(lower_w,upper_w))
```

#### Solution 2: Frequentist Test (two-sample test)

```python
# Create samples for black and white callbacks, respectively
b = data[data.race=='b'].call
w = data[data.race=='w'].call
```

```python
# z-test
z_stat,z_pval = ztest(b,w,alternative='two-sided')
print('z-statistic: {:4f}\np-value: {}'.format(abs(z_stat),z_pval))
```

```python
# t-test
t_stat, t_pval = stats.ttest_ind(b,w,equal_var=False)
print('z-statistic: {:4f}\np-value: {}'.format(abs(t_stat),t_pval))
```

The test statistic--for both the *z*-test and _t_-test--is 4.115, and the p-value is well below the threshold of $\alpha$=0.05. Thus, the null hypothesis must be rejected in favour of the alternative hypothesis at the 95% confidence interval.
